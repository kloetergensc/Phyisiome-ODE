```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
files = [
    1,2,3
]

TS = []
for f in files:
    TS.append(pd.read_parquet(f'../data/raw_data/electrophysiology/inada_N_2009/T=0.33_ds=0.1_dc=0.1/ts-data-{f}.parquet'))
   
data = TS[0].values
headers = TS[0].columns
n_chans = data.shape[-1] - 1
mnchans = min(n_chans, 200)
print(mnchans)
fig, axs = plt.subplots(mnchans, figsize=(20, 40))
for i in range(min(200,mnchans)):
    for j,ts in enumerate(TS):
        axs[i].plot(ts[headers[0]], ts[headers[i+1]],label = files[j])
    axs[i].set_ylabel(headers[i+1])
    axs[i].set_xlabel("t")
    axs[i].legend()
    
    # axs[i].set_title(f"Series {i}, channel {j+5}")
```
